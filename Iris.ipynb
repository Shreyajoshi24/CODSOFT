```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = pd.read_csv('IRIS.csv')
iris.columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Species']
iris.sample(5)
```

```python
encoder = LabelEncoder()
iris['Species_Code'] = encoder.fit_transform(iris['Species'])

features = iris[['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']]
target = iris['Species_Code']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='BuPu', xticklabels=encoder.classes_, yticklabels=encoder.classes_)
plt.title('Petal Prediction Matrix')
plt.xlabel('Predicted Species')
plt.ylabel('Actual Species')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

iris = pd.read_csv("IRIS.csv")
iris.columns = [col.strip().capitalize().replace(" ", "_") for col in iris.columns]
features = iris.drop("Species", axis=1)
target = iris["Species"]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
palette = {
    "Iris-setosa": "#6A9C89",
    "Iris-versicolor": "#C499F3",
    "Iris-virginica": "#F7A072"
}

pca_2d = PCA(n_components=2)
pca_2d_result = pca_2d.fit_transform(scaled_features)
explained_2d = pca_2d.explained_variance_ratio_

pca_2d_df = pd.DataFrame(pca_2d_result, columns=["PC1", "PC2"])
pca_2d_df["Species"] = target

sns.set(style="white", context="talk")

plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=pca_2d_df,
    x="PC1", y="PC2",
    hue="Species",
    palette=palette,
    s=150,
    edgecolor="black",
    linewidth=0.8,
    alpha=0.9
)
plt.title(f"2D PCA of IRIS Dataset", fontsize=18, fontweight='bold')
plt.xlabel(f"PC1 ({explained_2d[0]*100:.2f}%)", fontsize=14)
plt.ylabel(f"PC2 ({explained_2d[1]*100:.2f}%)", fontsize=14)
plt.legend(title="Species", fontsize=12, title_fontsize=13)
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()

pca_3d = PCA(n_components=3)
pca_3d_result = pca_3d.fit_transform(scaled_features)
explained_3d = pca_3d.explained_variance_ratio_

pca_3d_df = pd.DataFrame(pca_3d_result, columns=["PC1", "PC2", "PC3"])
pca_3d_df["Species"] = target

fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('#FAFAFA')
fig.patch.set_facecolor('#FAFAFA')

for species in iris["Species"].unique():
    subset = pca_3d_df[pca_3d_df["Species"] == species]
    ax.scatter(
        subset["PC1"], subset["PC2"], subset["PC3"],
        s=100,
        label=species,
        alpha=0.85,
        edgecolors='black',
        color=palette[species]
    )

ax.set_title("3D PCA Projection of IRIS Dataset", fontsize=18, fontweight='bold')
ax.set_xlabel(f"PC1 ({explained_3d[0]*100:.2f}%)", fontsize=12)
ax.set_ylabel(f"PC2 ({explained_3d[1]*100:.2f}%)", fontsize=12)
ax.set_zlabel(f"PC3 ({explained_3d[2]*100:.2f}%)", fontsize=12)
ax.tick_params(axis='both', labelsize=10)
ax.legend(title="Species", fontsize=11, title_fontsize=12)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

By performing PCA and visualizing in both 2D and 3D:

- I achieved insightful data compression, retaining most of the variance in fewer dimensions.

- The visualization clearly illustrates class separability, especially between Setosa and the other two species.

- This analysis forms a solid foundation for further modeling, clustering, or classification tasks.

```python
plt.figure(figsize=(8, 5))
pca_full = PCA()
pca_full.fit(scaled_features)
var_ratio = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(var_ratio)

plt.bar(range(1, len(var_ratio) + 1), var_ratio, alpha=0.6, color="#6A67CE", label='Individual Explained Variance')
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', color='black', label='Cumulative Explained Variance', linewidth=2)
plt.xticks(range(1, len(var_ratio) + 1))
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components', weight='bold')
plt.legend()
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()

scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_df["Species"] = target

sns.pairplot(
    scaled_df, 
    hue="Species", 
    palette=palette,
    diag_kind="kde",
    corner=True,
    plot_kws={"alpha": 0.7, "s": 60, "edgecolor": "white"}
)
plt.suptitle("Pairwise Relationships of Scaled Features", fontsize=16, y=1.02, weight='bold')
plt.tight_layout()
plt.show()
```
